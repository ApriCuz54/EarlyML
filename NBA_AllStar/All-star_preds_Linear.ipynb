```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/Users/apricuz/Data/2021-2022 NBA Player Stats - Regular.csv')
test = pd.read_csv('/Users/apricuz/Data/2022-2023 NBA Player Stats - Regular.csv')
```

```python
train.columns
```

```python
predictors = ['MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train[predictors], train['AS'])
predictions = model.predict(test[predictors])
```

```python
predictions
```

```python
predictions[0]
```

```python
for i in range(predictions.shape[0]):
    if predictions[i] > 0.3:
        predictions[i] = 1
    else:
        predictions[i] = 0
```

```python
predictions
```

```python
predictions = pd.DataFrame(predictions, columns = ["predictions"])
```

```python
predictions
```

```python
combination = pd.concat([test[["Player", "PTS", "AS"]], predictions], axis=1)
```

```python
combination
```

```python
combination.sort_values("predictions", ascending=False).head(49)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print('mean squared error: ', mean_squared_error(test['AS'], predictions))
print('mean absolute error: ', mean_absolute_error(test['AS'], predictions))
```

```python
combination.to_csv('result.csv', index = False)
```
